the catagorical features sex,CP,FBS,restecg,exang,slope,CA ,thaL

```python
sex_dmy = pd.get_dummies(heartdata['sex'], prefix='sex',drop_first=True)
cp_dmy = pd.get_dummies(heartdata['cp'],prefix='cp', drop_first=True)
fbs_dmy = pd.get_dummies(heartdata['fbs'],prefix='fbs', drop_first=True)
restecg_dmy = pd.get_dummies(heartdata['restecg'],prefix='restecg', drop_first=True)
exang_dmy = pd.get_dummies(heartdata['exang'],prefix='exang', drop_first=True)
slope_dmy = pd.get_dummies(heartdata['slope'],prefix='slope', drop_first=True)
ca_dmy = pd.get_dummies(heartdata['ca'],prefix='ca', drop_first=True)
thal_dmy = pd.get_dummies(heartdata['thal'],prefix='thal', drop_first=True)
```

```python
data = pd.concat([heartdata,sex_dmy,cp_dmy,fbs_dmy,restecg_dmy,exang_dmy,slope_dmy,ca_dmy,thal_dmy], axis=1)
```

```python
finaldata = data.drop(['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal','ID'],axis=1)
```

```python
finaldata
```

```python
finaldata.describe()
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
```

```python
feature_scale = ['age','trestbps','chol','thalach','oldpeak']
```

```python
finaldata[feature_scale] = standardScaler.fit_transform(finaldata[feature_scale])
```

```python
finaldata
```

```python
x= finaldata.drop(['target'],axis=1)
```

```python
y = finaldata['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_param = {
    'DecisionTreeClassifier':{
        'model':DecisionTreeClassifier(),
        'param':{
            'criterion': ['gini','entropy']
        }
    },
        'RandomForestClassifier':{
        'model':RandomForestClassifier(),
        'param':{
            'n_estimators': [20,50,80,120,150]
        }
    },
        'KNeighborsClassifier':{
        'model':KNeighborsClassifier(),
        'param':{
            'n_neighbors': [5,10,15,20,25]
        }
    },
     
}
```

```python
scores =[]
for model_name, mp in model_param.items():
    model_selection = GridSearchCV(estimator=mp['model'],param_grid=mp['param'],cv=5,return_train_score=False)
    model_selection.fit(x,y)
    scores.append({
        'model': model_name,
        'best_score': model_selection.best_score_,
        'best_params': model_selection.best_params_
    })
```

```python
df_model_score = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df_model_score
```

```python
model_rfc = RandomForestClassifier(n_estimators = 120)
```
